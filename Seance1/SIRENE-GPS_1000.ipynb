# Fonction pour transformation en chaîne de caractère

```python
def f(v, fin = False):
    r = ""
    if not(pandas.isna(v)):
        r += str(v)
        if not(fin):
            r += " "
    return r
```

# Importation des données ("step" premières lignes)

```python
import pandas # "pip install pandas" in terminal

fichier = "StockEtablissement_utf8_1000.csv"

step = 10
df = pandas.read_csv(fichier, nrows = step)
```

# Création de la colonne adresse

```python
adresse = [
    f(df["numeroVoieEtablissement"][i]) + f(df["indiceRepetitionEtablissement"][i]) +
    f(df["typeVoieEtablissement"][i]) + f(df["libelleVoieEtablissement"][i]) +
    f(df["codePostalEtablissement"][i]) + f(df["libelleCommuneEtablissement"][i], True) for i in range(df.shape[0])
    ]
```

# Création du fichier à envoyer

```python
fichier_a_envoyer = pandas.DataFrame({
    "siren": df["siren"],
    "siret": df["siret"],
    "adresse": adresse,
    "postcode": df["codePostalEtablissement"],
    "city": df["libelleCommuneEtablissement"]
})
fichier_a_envoyer.to_csv("export_for_search.csv", index = False)
```

# Envoi du fichier pour récupération des coordoonnées GPS

```python
import requests # "pip install requests" in terminal

url = "https://api-adresse.data.gouv.fr/search/csv/"
fichier = open("export_for_search.csv", "rb") # ouverture en lecture ("r") binaire ("b")

r = requests.post(url, files = {"data": fichier})
```

# Ecriture de ces résultats dans un fichier csv pour relecture ensuite

```python
fichier_resultat = open("export_for_search_results.csv", "w", encoding = "utf-8")
fichier_resultat.write(r.text)
fichier_resultat.close()
```

# Boucle

```python
start = 0
end = False
while(start<end):
    # Code à exécuter pour récupérer les longitudes et latitudes de chaque entreprise
    for i in range(fichier.shape[0]):
        a = fichier.loc[i,:]
        # Création de la colonne adresse
        adresse = f(a["numeroVoieEtablissement"]) + f(a["indiceRepetitionEtablissement"]) +\
            f(a["typeVoieEtablissement"]) + f(a["libelleVoieEtablissement"]) +\
            f(a["codePostalEtablissement"]) + f(a["libelleCommuneEtablissement"], True)
            
        recherche = url + "?q=" + adresse
        reponse = requests.get(recherche)
        coord = reponse.json()["features"][0]["geometry"]["coordinates"]
        fichier.loc[i,"lng"] = coord[0]
        fichier.loc[i,"lat"] = coord[1]

    # Affichage des résultats
    print(fichier[["siren", "lng", "lat"]])

    # Export
    fichier.to_csv("export_for_search.csv", index = False)
    import requests

    url = "https://api-adresse.data.gouv.fr/search/csv/"
    fichier = open("export_for_search.csv", "rb") # ouverture en lecture ("r") binaire ("b")

    r = requests.post(url, files = {"data": fichier})
    r.text # contenu du résultat dans un format CSV
    fichier_resultat = open("export_for_search_results.csv", "w", encoding = "utf-8")
    fichier_resultat.write(r.text)
    fichier_resultat.close()
    pandas.read_csv("export_for_search_results.csv")
    df_ajout = pandas.concat([df, resultats.filter(["latitude","longitude"])])
    if start == 0:
        df_ajout.to_csv("export_for_search.csv", index = False)
    else:
        df_ajout.to_csv("export_for_search.csv", index = False, header = False, mode = "a")
    start += step
```
